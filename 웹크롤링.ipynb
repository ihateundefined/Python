```python
# 25.04.22
from os import link
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

page = requests.get('https://library.gabia.com/')
soup = bs(page.text, "html.parser")

ele = soup.select("div.esg-entry-content a.eg-grant-element-0")

titles = []
links = []

for index, e in enumerate(ele) :
  titles.append(e.text)
  links.append(e.attrs["href"])

df = pd.DataFrame()
# 데이터 전처리

df[titles] = titles
df[links] = links

df.to_csv("gabia.csv", index=False)
# df.to_excel("gabia.xlsx", sheet_name = "gabia")
```
