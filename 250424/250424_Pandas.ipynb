# Pandas

```python
import pandas as pd

# 1차원 배열 Vector의 Series
numSeries = pd.Series([1,2,3,4,5])
print(numSeries)

numSeries2 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(numSeries2)
```

```python
import pandas as pd

# 2차원 배열 Matrix의 DataFrame
print('field가 1개')
df = pd.DataFrame([1,2,3,4,5])
print(df)


print('field가 3개')
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

datas = pd.date_range('today', periods=6)
numSeries = pd.Series([100,20,300,80,500,600], index=datas)

print(numSeries)
numSeries.plot()
plt.show()

numSeries.plot(kind='bar', stacked=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

num_df = pd.DataFrame({'A':{'a':10, 'b':20, 'c':30, 'd':40},
                       'B':{'a':5, 'b':10, 'c':15, 'd':20}})

print(num_df)
num_df.plot(kind='bar', stacked=True)
plt.show()

num_df.plot()
plt.show()
```
