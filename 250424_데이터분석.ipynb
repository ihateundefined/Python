```python
import csv

f = open('seoul.csv', 'r', encoding='utf-8')
data = csv.reader(f)

# header = next(data) # 파일의 첫번째 행 skip

next(data)

julyMax = []
julyLow = []
augustMax = []
augustLow = []
myBirthdayMax = []
myBirthdayLow = []

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '07':
     julyMax.append(float(row[-1]))
     julyLow.append(float(row[-2]))
    if row[0].split('-')[1] == '08':
     augustMax.append(float(row[-1]))
     augustLow.append(float(row[-2]))
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '27':
      myBirthdayMax.append(float(row[-1]))
      myBirthdayLow.append(float(row[-2]))

f.close()

# print(result)
plt.plot(julyMax, 'hotpink', label='max tempt')
plt.plot(julyLow, 'lightgray', label='lowest tempt')
plt.title('july')
plt.legend()
plt.show()

plt.plot(augustMax, 'hotpink', label='max tempt')
plt.plot(augustLow, 'lightgray', label='lowest tempt')
plt.title('august')
plt.legend()
plt.show()

plt.plot(myBirthdayMax, 'hotpink', label='max tempt')
plt.plot(myBirthdayLow, 'lightgray', label='lowest tempt')
plt.title('my birthday')
plt.legend()
plt.show()
```

```python
import csv

f = open('seoul.csv', 'r', encoding='utf-8')
data = csv.reader(f)

next(data)

result = []
january = []
august = []

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))
    if row[0].split('-')[1] == '01':
      january.append(float(row[-1]))
    if row[0].split('-')[1] == '08':
      august.append(float(row[-1]))

f.close()

plt.hist(result, bins=100, color='yellowgreen')
plt.hist(january, bins=100, color='skyblue')
plt.hist(august, bins=100, color='hotpink')
plt.title('all')
plt.show()
```

# CSV 파일 (연령) 시각화

1. 인구 데이터 파일 읽기
2. 전체 데이터를 한 줄씩 반복해서 읽기
3. 알고 싶은 동네 데이터 찾기
4. 0세부터 100세까지 인구수를 순서대로 저장
5. 연령별 인구수 데이터를 시각화

```python
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

# print(data) # data는 object iterator 상태라서 for문 반드시 사용해줘야함
for row in data:
  if '4777035000' in row[0]: # 영덕군 축산면
    for i in row[3:]:
      result.append(int(i))

# print(result)
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

dongne = input('찾고 싶은 지역의 동네 이름을 입력하세요 : ')

for row in data:
  if dongne in row[0]:
    for i in row[3:]:
      result.append(int(i))

# print(result)
plt.style.use('ggplot')
plt.bar(range(101), result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

male = []
female = []

dongne = input('찾고 싶은 지역의 동네 이름을 입력하세요 : ')

for row in data:
  if dongne in row[0]:
    for i in range(1, 101):
      male.append(int(row[i+3]))
      female.append(int(row[-(i+1)]))
female.reverse()

# print('남성', male)
# print('여성', female)

plt.style.use('ggplot')
plt.plot(male, 'hotpink', label='male')
plt.plot(female, 'skyblue', label='female')
plt.legend()
plt.show()

# vertical plot
plt.style.use('ggplot')
plt.bar(range(100), male, color='hotpink', label='male')
plt.bar(range(100), female, color='skyblue', label='female')
plt.legend()
plt.show()

# horizontal plot
plt.style.use('ggplot')
plt.barh(range(100), male, color='hotpink', label='male')
plt.barh(range(100), female, color='skyblue', label='female')
plt.legend()
plt.show()
```
